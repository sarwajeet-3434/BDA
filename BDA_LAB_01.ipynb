```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import drive
# Mount Google Drive
drive.mount('/content/drive')

# Load the dataset
df = pd.read_csv('/content/drive/My Drive/BDA/Dataset/bank_marketing_updated_v1.csv', header=2)

# Display the first few rows
df.head()
```

```python
#Unvariate Analysis (categorical variable)
# Select categorical columns
categorical_cols = df.select_dtypes(include='object').columns

# Plot distribution of each categorical column
for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=df, x=col, palette='Set2')
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
#Unvariate Analysis Numerical Variable
# Select numerical columns
numerical_cols = df.select_dtypes(include=np.number).columns

# Plot distribution of each numerical column
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True, color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.tight_layout()
    plt.show()
```

```python
# Bivariate Analysis - Categorical Variables
for col in categorical_cols:
    if col != 'response':  # Avoid plotting target against itself
        plt.figure(figsize=(8, 4))
        sns.countplot(data=df, x=col, hue='response', palette='Set1')
        plt.title(f'{col} vs Term Deposit Subscription')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
```

```python
#Bivariate Ananlysis (Numerical Variables)
df['response'] = df['response'].astype(str)
sampled_df = df.sample(n=500, random_state=42)

# Select only numeric columns
numeric_cols = sampled_df.select_dtypes(include=np.number).columns.tolist()

# Generate pairplot
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(sampled_df[numeric_cols + ['response']], hue='response', palette='husl')
plt.suptitle("Pairplot of Numerical Variables by Response", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
corr_matrix = df[numeric_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Numerical Features")
plt.show()
```
